```python
import numpy as np
import matplotlib.pyplot as plt
```

## Normal Distribution
###### The following is a normal distribution with $\mu = 0, \sigma=2$.

```python
mean = 0
sigma = 2
normal_distribution = np.random.normal(mean, sigma, 100)
x = np.linspace(-10, 10, 100)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x-mean)**2 / (2 * sigma **2)), color='r')
plt.grid()
plt.title(r'Normal distribution for $\mu=0, \sigma=2$')
plt.show()
```

## Monte-Carlo Simulation
###### The following code-snippet is a monte carlo simulation for different sized samples with $\mu = 4,\sigma=4\sqrt{2}$. It is found out that 1000 simulations were enough to get a bell like structure. In 10000 simulations, the result didn't differ a lot than 1000 simulations.

```python
sizes = [50, 100, 1000, 10000]
mean = 4
sigma = 4 * np.sqrt(2)
for i in range(len(sizes)):
    x = np.random.normal(mean, sigma, sizes[i])
    plt.figure(dpi=120)
    plt.subplot(2, 2, i+1)
    plt.hist(x, bins=20)
    plt.title('%d samples' % sizes[i])
    plt.xticks([])
plt.show()   
```
